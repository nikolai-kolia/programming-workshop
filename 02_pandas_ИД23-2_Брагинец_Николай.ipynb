### Группировки таблиц `pd.DataFrame`

4.1 Посчитайте количество рецептов, представленных каждым из участников (`contributor_id`). Какой участник добавил максимальное кол-во рецептов?

```python
# Группировка рецептов по идентификатору автора и подсчет количества рецептов каждым автором
counts_by_id = recipes.groupby('contributor_id')['name'].count().reset_index(name = 'num_recipes')

# Сортировка выходного датафрейма по количеству рецептов каждого автора в порядке убывания
counts_by_id.sort_values('num_recipes', ascending=False)
```

4.2 Посчитайте средний рейтинг к каждому из рецептов. Для скольких рецептов отсутствуют отзывы? Обратите внимание, что отзыв с нулевым рейтингом или не заполненным текстовым описанием не считается отсутствующим.

```python
# Группировка отзывов по идентификатору рецепта и подсчет среднего рейтинга каждого рецепта
srednee = reviews.groupby('recipe_id')['rating'].mean().reset_index(name = 'mean')

# Группировка отзывов по идентификатору рецепта и подсчет количества отзывов каждому рецепту
amount = reviews.groupby('recipe_id')['rating'].count().reset_index(name = 'amount')

# Слияние двух выходных датафреймов по идентификатору рецепта
merged = srednee.join(amount['amount'])

# Переименование столбца 'Unnamed: 0' в 'something' в датафрейме 'reviews'
reviews = reviews.rename(columns = {'Unnamed: 0': 'something'})

# Группировка отзывов с пропущенным значением в поле 'review' по идентификатору рецепта и подсчет количества таких отзывов
NaNs = reviews[reviews['review'].isnull()].groupby('recipe_id')['review'].apply(lambda x: len(x)).reset_index(name = 'amount_of_NaN')

# Вывод датафрейма 'NaNs'
NaNs
```

4.3 Посчитайте количество рецептов с разбивкой по годам создания.

```python
# Преобразование столбца 'date' в формат даты и времени
reviews['date'] = pd.to_datetime(reviews['date'])

# Сортировка отзывов по дате
corrected_reviews = reviews.sort_values('date')

# Извлечение года из столбца 'date'
corrected_reviews['date'] = corrected_reviews['date'].dt.year

# Группировка отзывов по году и подсчет количества отзывов в каждом году
year_freq = reviews.groupby('date')['review'].count().reset_index(name = 'freq')

# Сортировка частоты отзывов по годам в порядке убывания
year_freq.sort_values('freq', ascending=False)
```

5.1 При помощи объединения таблиц, создайте `DataFrame`, состоящий из четырех столбцов: `id`, `name`, `user_id`, `rating`. Рецепты, на которые не оставлен ни один отзыв, должны отсутствовать в полученной таблице. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и попытавшись найти строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
# Объединение датафреймов 'recipes' и 'reviews' по идентификатору рецепта и идентификатору отзыва
merged = pd.merge(recipes, reviews, left_on='id', right_on='recipe_id')

# Отбор строк, где отзывы не являются пустыми
merged = merged[merged['review'].notnull()]

# Выбор необходимых столбцов 'id', 'name', 'user_id', 'rating' и сортировка по идентификатору рецепта
desired_columns = ['id', 'name', 'user_id', 'rating']
merged = merged[desired_columns].sort_values('id')

# Вывод результирующего датафрейма 'merged'
merged
```

5.2 При помощи объединения таблиц и группировок, создайте `DataFrame`, состоящий из трех столбцов: `recipe_id`, `name`, `review_count`, где столбец `review_count` содержит кол-во отзывов, оставленных на рецепт `recipe_id`. У рецептов, на которые не оставлен ни один отзыв, в столбце `review_count` должен быть указан 0. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и найдя строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
# Объединение датафреймов 'reviews' и 'recipes' по идентификатору рецепта и идентификатору рецепта
connected_dataframe = pd.merge(reviews, recipes, left_on='recipe_id', right_on='id').sort_values('recipe_id')

# Выбор необходимых столбцов 'recipe_id', 'name', 'review_count'
desired_columns2 = ['recipe_id', 'name', 'review_count']

# Отбор строк, где отзывы являются пустыми, и вывод необходимых столбцов
nulls = connected_dataframe[connected_dataframe['review'].isnull()][desired_columns2[:-1]]

# Установка значения 'review_count' в 0 для строк с пустыми отзывами
nulls['review_count'] = 0

# Группировка отзывов по идентификатору рецепта и подсчет количества отзывов каждому рецепту
review_count = connected_dataframe.groupby('recipe_id')['review'].count().reset_index(name = 'review_count')

# Объединение датафреймов 'connected_dataframe' и 'review_count' по идентификатору рецепта
connected_dataframe_with_review_count = pd.merge(connected_dataframe, review_count, left_on = 'recipe_id', right_on = 'recipe_id')

# Вывод результирующего датафрейма 'connected_dataframe_with_review_count' с необходимыми столбцами и сортировка по количеству отзывов
connected_dataframe_with_review_count = connected_dataframe_with_review_count[desired_columns2]._append(nulls).sort_values('review_count').drop_duplicates()

# Вывод результирующего датафрейма 'connected_dataframe_with_review_count'
connected_dataframe_with_review_count
```

5.3. Выясните, рецепты, добавленные в каком году, имеют наименьший средний рейтинг?

```python
# Извлечение года из столбца 'submitted'
connected_dataframe['submitted'] = connected_dataframe['submitted'].dt.year

# Группировка данных по году подачи рецепта и подсчет среднего рейтинга каждого года
mean_rating = connected_dataframe.groupby('submitted')['rating'].mean().reset_index(name = 'mean_rating')

# Переименование столбца 'submitted' в 'year_of_recipe'
mean_rating.rename(columns = {'submitted': 'year_of_recipe'})

# Сортировка выходного датафрейма по среднему рейтингу в порядке убывания
mean_rating.sort_values('mean_rating')
```
