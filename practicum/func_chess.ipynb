```python
# Шахматы с дополнениями

board = [
    ['r', 'n', 'b', 'q', 'k', 'b', 'n', 'r'],
    ['p', 'p', 'p', 'p', 'p', 'p', 'p', 'p'],
    ['.', '.', '.', '.', '.', '.', '.', '.'],
    ['.', '.', '.', '.', '.', '.', '.', '.'],
    ['.', '.', '.', '.', '.', '.', '.', '.'],
    ['.', '.', '.', '.', '.', '.', '.', '.'],
    ['P', 'P', 'P', 'P', 'P', 'P', 'P', 'P'],
    ['R', 'N', 'B', 'Q', 'K', 'B', 'N', 'R'],
]

steps = 0


def print_board(board):
    """
    Функция для вывода шахматной доски в консоль.
    Принимает доску в виде двумерного списка.
    Выводит координаты по горизонтали (a-h) и вертикали (1-8),
    а также содержимое каждой клетки доски
    """
    print("  a b c d e f g h")
    for i in range(8):
        print(1 + i, end="|")
        for j in range(8):
            print(board[i][j], end=" ")
        print("|")
    print("  a b c d e f g h")


def do_step(board, start, end):
    """
    Функция для выполнения хода на шахматной доске.
    Принимает доску, координаты начальной и конечной клеток.
    Перемещает фигуру с начальной клетки на конечную.
    Также увеличивает счетчик ходов
    """
    global steps
    piece = board[start[0]][start[1]]
    board[start[0]][start[1]] = '.'
    board[end[0]][end[1]] = piece
    steps += 1


def is_valid_step(board, start, end):
    """
    Функция для проверки валидности хода на шахматной доске.
    Принимает доску, координаты начальной и конечной клеток.
    Возвращает True, если ход валидный, иначе False.
    Учитывает правила передвижения всех фигур
    """
    piece = board[start[0]][start[1]]
    if piece == '.':
        return False
    # Проверка хода для пешки
    if board[end[0]][end[1]] == '.':
        if piece.isupper():
            if start[0] == 6 and start[0] - end[0] == 2 and start[1] == end[1]:
                return True
            elif start[0] - end[0] == 1 and start[1] == end[1]:
                return True
        else:
            if start[0] == 1 and end[0] - start[0] == 2 and start[1] == end[1]:
                return True
            elif end[0] - start[0] == 1 and start[1] == end[1]:
                return True
    # Проверка для взятия фигуры c ДОП УСЛОВИЕМ 9
    if piece.isupper() != board[end[0]][end[1]].isupper():
        if piece == 'P' and start[0] - end[0] == 1 and abs(start[1] - end[1]) == 1:
            if (start[0], start[1]) in board.en_passant_captures and board.en_passant_captures[(start[0], start[1])] == (end[0], end[1]):
                return True
            return False
        elif piece == 'p' and end[0] - start[0] == 1 and abs(start[1] - end[1]) == 1:
            if (start[0], start[1]) in board.en_passant_captures and board.en_passant_captures[(start[0], start[1])] == (end[0], end[1]):
                return True
            return False
    if (piece.islower() and end[0] == 0) or (piece.isupper() and end[0] == 7):
        return True
    return False

    # Далее проверяются ходы для остальных фигур
    if piece == '.':
        return False
    if board[end[0]][end[1]] == '.':
        if piece.isupper():
            # Проверка хода для ладьи (для белых фигур)
            if piece == 'R' or piece == 'r':
                if start[0] == end[0] or start[1] == end[1]:
                    if start[0] == end[0]:
                        for i in range(min(start[1], end[1]) + 1, max(start[1], end[1])):
                            if board[start[0]][i] != '.':
                                return False
                    else:
                        for i in range(min(start[0], end[0]) + 1, max(start[0], end[0])):
                            if board[i][start[1]] != '.':
                                return False
                    return True
            # Проверка хода для коня (для белых фигур)
            elif piece == 'N' or piece == 'n':
                if abs(start[0] - end[0]) == 2 and abs(start[1] - end[1]) == 1:
                    return True
                elif abs(start[0] - end[0]) == 1 and abs(start[1] - end[1]) == 2:
                    return True
            # Проверка хода для слона (для белых фигур)
            elif piece == 'B' or piece == 'b':
                if abs(start[0] - end[0]) == abs(start[1] - end[1]):
                    row_dir = 1 if end[0] - start[0] > 0 else -1
                    col_dir = 1 if end[1] - start[1] > 0 else -1
                    for i in range(1, abs(start[0] - end[0])):
                        if board[start[0] + (i * row_dir)][start[1] + (i * col_dir)] != '.':
                            return False
                    return True
        else:
            # Проверка хода для ладьи (для черных фигур)
            if piece == 'r' or piece == 'r':
                if start[0] == end[0] or start[1] == end[1]:
                    if start[0] == end[0]:
                        for i in range(min(start[1], end[1]) + 1, max(start[1], end[1])):
                            if board[start[0]][i] != '.':
                                return False
                    else:
                        for i in range(min(start[0], end[0]) + 1, max(start[0], end[0])):
                            if board[i][start[1]] != '.':
                                return False
                    return True
            # Проверка хода для коня (для черных фигур)
            elif piece == 'n' or piece == 'n':
                if abs(start[0] - end[0]) == 2 and abs(start[1] - end[1]) == 1:
                    return True
                elif abs(start[0] - end[0]) == 1 and abs(start[1] - end[1]) == 2:
                    return True
            # Проверка хода для слона (для черных фигур)
            elif piece == 'b' or piece == 'b':
                if abs(start[0] - end[0]) == abs(start[1] - end[1]):
                    row_dir = 1 if end[0] - start[0] > 0 else -1
                    col_dir = 1 if end[1] - start[1] > 0 else -1
                    for i in range(1, abs(start[0] - end[0])):
                        if board[start[0] + (i * row_dir)][start[1] + (i * col_dir)] != '.':
                            return False
                    return True
    # Проверка для взятия фигуры (для ладьи, коня и слона)
    if piece.isupper() != board[end[0]][end[1]].isupper():
        if piece == 'R' or piece == 'r':
            if start[0] == end[0] or start[1] == end[1]:
                if start[0] == end[0]:
                    for i in range(min(start[1], end[1]) + 1, max(start[1], end[1])):
                        if board[start[0]][i] != '.':
                            return False
                else:
                    for i in range(min(start[0], end[0]) + 1, max(start[0], end[0])):
                        if board[i][start[1]] != '.':
                            return False
                return True
        elif piece == 'N' or piece == 'n':
            if abs(start[0] - end[0]) == 2 and abs(start[1] - end[1]) == 1:
                return True
            elif abs(start[0] - end[0]) == 1 and abs(start[1] - end[1]) == 2:
                return True
        elif piece == 'B' or piece == 'b':
            if abs(start[0] - end[0]) == abs(start[1] - end[1]):
                row_dir = 1 if end[0] - start[0] > 0 else -1
                col_dir = 1 if end[1] - start[1] > 0 else -1
                for i in range(1, abs(start[0] - end[0])):
                    if board[start[0] + (i * row_dir)][start[1] + (i * col_dir)] != '.':
                        return False
                return True

    # Проверка хода для ферзя
    if piece.upper() != 'Q':
        return False
    if start[0] == end[0] or start[1] == end[1] or abs(start[0] - end[0]) == abs(start[1] - end[1]):
        if start[0] == end[0]:
            for i in range(min(start[1], end[1]) + 1, max(start[1], end[1])):
                if board[start[0]][i] != '.':
                    return False
        elif start[1] == end[1]:
            for i in range(min(start[0], end[0]) + 1, max(start[0], end[0])):
                if board[i][start[1]] != '.':
                    return False
        else:
            row_dir = abs(start[0] - end[0])
            col_dir = abs(start[1] - end[1])
            for i in range(1, row_dir):
                if start[0] < end[0]:
                    row = start[0] + i
                else:
                    row = start[0] - i
                if start[1] < end[1]:
                    col = start[1] + i
                else:
                    col = start[1] - i
                if board[row][col] != '.':
                    return False
        return True

    # Проверка хода для короля
    piece = board[start[0]][start[1]]
    if piece.upper() == 'K':
        if abs(start[0] - end[0]) == 2 and abs(start[1] - end[1]) == 0:
            # Castling move
            rook_start = (start[0], start[1] + (-1 if end[0] > start[0] else 1))
            rook_end = (start[0], end[1])
            if board[rook_start[0]][rook_start[1]] == 'R' and board[rook_end[0]][rook_end[1]] == '.' and all(board[i][start[1] + (-1 if end[0] > start[0] else 1)] == '.' for i in range(start[0] + (-1 if end[0] > start[0] else 1), end[0])):
                return True
            
    move_history.append((start, end))

def rollback_move(board, move_history, num_moves=1):
    """
    Функция для отката заданного количества ходов на шахматной доске.
    Берет доску, историю ходов и количество ходов для отката.
    Обновляет состояние доски, отменяя указанные ходы.
    """
    for _ in range(num_moves):
        if move_history:
            last_move = move_history.pop()
    
    
# Функция для игры
def play_game():
    turn = 0

    while True:
        print_board(board)
        if turn % 2 == 0:
            print("Ход белых:")
        else:
            print("Ход черных:")
        start = input("Введите начальное положение фигуры (например, e2): ")
        end = input("Введите конечное положение фигуры (например, e4): ")
        start_row = int(start[1]) - 1
        start_col = ord(start[0]) - ord('a')
        end_row = int(end[1]) - 1
        end_col = ord(end[0]) - ord('a')
        if is_valid_step(board, (start_row, start_col), (end_row, end_col)):
            do_step(board, (start_row, start_col), (end_row, end_col))
            turn += 1
        else:
            print(f"\n\nНекорректный ход. Попробуйте еще раз.")
        print(f"Кол-во сделанных ходов {steps}")


# Запуск игры
play_game()
```
